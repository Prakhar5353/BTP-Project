```python
import pandas as pd
```

```python
df_dph = pd.read_csv('signal_200.csv')
df_dph['class'] = [1]*10000
df_dph['weight'] =  469.2*1000*100/10000
df_dph#let us consider a luminosity of 100fb-1 
#df_dph = df_dph.sample(frac=1)
```

```python
df_bkg = pd.read_csv('background.csv')
df_bkg['class'] = [0]*10000
df_bkg['weight'] =  0.5157*1000*100/10000
df_bkg#let us consider a luminosity of 100fb-1 
#df_dph = df_dph.sample(frac=1)
```

```python
df = pd.concat([df_bkg,df_dph],axis = 0, join = 'outer')
df = df.sample(frac = 1)
df
```

```python
#from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
import matplotlib.pyplot as plt
import scienceplots
import scienceplots
plt.style.use(['science', 'notebook'])
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np
```

```python
target = df['class']
features = df[['px','py','pz','E','pt','cosphi','mis_e','eta']]
weights = np.array(df['weight'])
```

```python
cols = features.columns
df_scaled = StandardScaler().fit(features)##important
dfsc = df_scaled.transform(features)
scaled_features = pd.DataFrame(dfsc,
                         columns=cols)
scaled_features.describe()
```

```python
scaled_features['train_wt'] = weights
```

```python
class CustomDataset(Dataset):
    def __init__(self, inputs, targets,weights):
        self.inputs = inputs
        self.targets = targets
        self.weights = weights

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx], self.targets[idx], self.weights[idx]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features, target,test_size = 0.3, random_state = 0)
X_train.describe()
```

```python
X_test = X_test.drop('train_wt',axis=1)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.0625)
X_train.describe()
```

```python
X_train_features, X_train_weight = X_train.drop('train_wt',axis=1), X_train['train_wt']
X_valid_features, X_valid_weight = X_valid.drop('train_wt',axis=1), X_valid['train_wt']
```

```python
training_inputs,training_weights, training_outputs = torch.tensor(X_train_features.values, dtype = torch.float).to(device),torch.tensor(X_train_weight.values, dtype = torch.float).to(device),torch.tensor(y_train.values, dtype = torch.float).reshape(-1, 1).to(device)
validation_inputs,validation_weights, validation_outputs = torch.tensor(X_valid_features.values, dtype = torch.float).to(device),torch.tensor(X_valid_features.values, dtype = torch.float).to(device),torch.tensor(y_valid.values, dtype = torch.float).reshape(-1, 1).to(device)
#training_outputs.reshape(-1, 1).shape
```

```python
train_dataset = CustomDataset(training_inputs, training_outputs,training_weights)
train_dataloader = DataLoader(train_dataset, batch_size=512)
```

```python
validation_dataset = CustomDataset(validation_inputs, validation_outputs,validation_weights)
validation_dataloader = DataLoader(validation_dataset, batch_size=100)
```

```python
X_train_features
```

```python
class classifier(nn.Module):
    def __init__(self,layer,input_dim):
        super().__init__()
        self.layer = layer
        self.input_dim = input_dim
        
        self.hiddenlayer1 = nn.Linear(in_features = self.input_dim, out_features = self.layer[0])
        self.activ1 = nn.ReLU()
        self.hiddenlayer2 = nn.Linear(in_features = self.layer[0],out_features = self.layer[1])
        self.activ2 = nn.ReLU()
        self.hiddenlayer3 = nn.Linear(in_features = self.layer[1],out_features = self.layer[2])
        self.activ3 = nn.ReLU()
        self.hiddenlayer4 = nn.Linear(in_features = self.layer[2],out_features = self.layer[3])
        self.activ4 = nn.ReLU()
        self.hiddenlayer5 = nn.Linear(in_features = self.layer[3],out_features = self.layer[4])
        self.activ5 = nn.ReLU()
        self.hiddenlayer6 = nn.Linear(in_features = self.layer[4],out_features = self.layer[5])
        self.activ6 = nn.Sigmoid()
    def forward(self, inputs):
        out1 = self.hiddenlayer1(inputs)
        out2 = self.activ1(out1)
        out3 = self.hiddenlayer2(out2)
        out4 = self.activ2(out3)
        out5 = self.hiddenlayer3(out4)
        out6 = self.activ3(out5)
        out7 = self.hiddenlayer4(out6)
        out8 = self.activ4(out7)
        out9 = self.hiddenlayer5(out8)
        out10 = self.activ5(out9)
        out11 = self.hiddenlayer6(out10)
        output = self.activ6(out11)
        return output 
```

```python
hidden_layers = [16,32,64,32,8,1]
model = classifier(hidden_layers, 8)
model.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
criterion = torch.nn.BCELoss(reduction = 'none')
n_epochs = 50
valid_loss = []
train_loss = []
for epoch in range(n_epochs):
        print(f"Epoch {epoch + 1}/{n_epochs}")
        t_loss = []
        for step, data in enumerate(train_dataloader):
            optimizer.zero_grad()
            prediction = model.forward(data[0])
            #print(prediction)
            #print("data: ",data[1])
            loss = torch.mean(criterion(prediction, data[1])*data[2])
            loss.backward()
            optimizer.step()
            t_loss.append(loss)
        train_loss.append(torch.tensor(t_loss).mean())
        
        v_loss = []
        for v_step, data in enumerate(validation_dataloader):
            with torch.no_grad():
                _prediction = model.forward(data[0])
                _loss = torch.mean(criterion(_prediction, data[1])*data[2])
                v_loss.append(_loss)
        valid_loss.append(torch.tensor(v_loss).mean())

        #torch.save(model.state_dict(), f'epoch_{epoch}valloss{valid_loss[epoch]:.3f}.pt')
        print(f"Loss = {train_loss[epoch]:.5f},\t val_loss = {valid_loss[epoch]:.5f}")
```

```python
test_input, test_output = torch.tensor(X_test.values, dtype = torch.float).to(device),torch.tensor(y_test.values, dtype = torch.float).reshape(-1,1).to(device)
```

```python
pred_test = model.forward(test_input)
pred_test.shape
```

```python
pred_test = pred_test.detach().cpu().numpy()
test_output = test_output.detach().cpu().numpy()
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(test_output, pred_test)
```

```python
from sklearn.metrics import auc
auc(fpr,tpr)
```

```python
plt.plot(fpr,tpr)
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.grid(True)
plt.title("ROC for classification with All Background")
plt.savefig("all_roc.pdf")
plt.show()
```
