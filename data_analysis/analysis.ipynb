```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook'])
```

```python
sig = pd.read_csv('signal.csv')
bkg = pd.read_csv('background.csv')
```

```python
sig_mis_e = np.array(sig['mis_e'])
bkg_mis_e = np.array(bkg['mis_e'])
```

```python
num = (np.max(sig_mis_e)-np.min(sig_mis_e))*2
print(num)
```

```python
plt.hist(sig_mis_e,bins=40,density = True, color = 'green', alpha = 0.75,label = "signal events")
plt.hist(bkg_mis_e,bins=40,density = True, color = 'red', alpha = 0.5,label = "background events")
plt.legend()
plt.grid(True)
plt.show()
```

```python
h1, b1 = np.histogram(sig_mis_e, bins = 40, range = (12, 24))
h2, b2 = np.histogram(bkg_mis_e, bins = 40, range = (12, 24))
```

```python
p1 = h1 / 10000 
p2 = h2 / 10000
```

```python
plt.hist(b1[1:], bins=b1, weights = p1, color = 'green', alpha = 0.75,label = "signal events")
plt.hist(b1[1:], bins=b1, weights = p2, color = 'red', alpha = 0.5,label = "background events")
plt.legend()
plt.grid(True)
plt.savefig('mis_e.png')
plt.show()
```
